```python
import matplotlib.pyplot as plt
```

```python
x = [50,42,39,34,32,48,22,17,8,13]
y=[17,19,21,22,24,23,34,31,43,35]
# first city had 50 rides with an average fare of $17
```

```python
plt.scatter(x,y, c = 'green', marker = 'x', s =x)
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Average Fare vs Total Rides')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
# When we use the plt.plot() function to create a scatter plot, #
# we need to add a lowercase “o” as a parameter inside the parentheses. 
# This switches the plot from a line chart to a scatter plot.
```

```python
plt.scatter(y_axis, x_axis, c = 'Red', label = 'Chicago')
plt.ylabel('Date')
plt.xlabel('Fare ($)')
plt.title('PyBer Fare by Month')
plt.gca().invert_yaxis()
plt.legend()
plt.xlim(0,45)
plt.show()
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
# change the size of our marker
```

```python
# makre the sizes larger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# You can use list comprehension to replace many for and while loops. 
# List comprehension is faster because it is optimized for Python to spot a predictable pattern during looping.
# new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "skyblue", s = [i * 5 for i in y_axis], label = 'Boston', alpha = .8, edgecolor='black',linewidth = 2.0)
ax.set_xlim(0,50)
ax.legend()
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")

ax.invert_yaxis()
ax.set_title("PyBer Fare By Month")
```
